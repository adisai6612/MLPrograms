```python
import pandas as pd
import numpy as np
df = pd.read_csv("C:\\Users\\HP\\Downloads\\Salary_Data.csv")
#print(df)
df.head()
#print(df.info())
```

```python
X = df.iloc[:, :-1].values   #:all rows column last 1 delete
Y = df.iloc[:,1].values

print(X)
print(Y)
```

```python
pip install matplotlib
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,Y,c="green")
plt.show()
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)

print('Shape of X_train =',X_train.shape)
print('Shape of Y_train =',Y_train.shape)
print('Shape of X_train =',X_test.shape)
print('Shape of Y_train =',Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,Y_train)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
print(X_test) #actual value
```

```python
lr.predict(X_test)#predicted value
```

```python
print(Y_test)
```

```python
Y_predict=lr.predict(X_test)
print(Y_predict)
```

```python
from sklearn.metrics import mean_absolute_error #error
m=mean_absolute_error(Y_test,Y_predict)
print(m)
```

```python
#Mean Absolute Error is used where outlier is present.
#THe sum of square is not applicable when outlier is present in the dataset

from sklearn.metrics import mean_squared_error #error
m=mean_squared_error(Y_test,Y_predict)
print(m)
```

```python

plt.scatter(Y_test, Y_predict, color="blue")
plt.title("Actual vs Predicted Values")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_train,Y_train,color='red')

plt.plot(X_train,lr.predict(X_train),color = 'green')
plt.title('Salary vs Experience (Traning Set)')
plt.xlabel('Year of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
inputexp=12
new_X=lr.predict([[inputexp]])
new_X
```

```python
#Visualising the test set results
plt.scatter(X_test,Y_test,color='red')

plt.plot(X_train,lr.predict(X_train),color = 'green')
plt.scatter(inputexp,new_X, color ='blue')
plt.title('Salary vs Experience (Traning Set)')
plt.xlabel('Year of Experience')
plt.ylabel('Salary')
plt.show()
```
